### Environment Setup & Model Initialization for Graph Coloring
This section contains the environment setup, helper functions, and initialization code for the graph coloring experiment, including path configuration, random seed control, and device setup.


First of all, ensure you can import our main code `src`:

```python
import sys
from pathlib import Path
current = Path.cwd()
root_path = current.parents[3]  
sys.path.append(str(root_path))
import time
import random
import numpy as np
import torch
import torch.nn as nn
from src.coloring import loss_coloring_onehot_qubo
from src import from_file_to_graph, init, get_device, run_qubo
from src import Layer, LayerType, Datasets
from src.coloring import  DirectProbModel
```

### Set random seeds for reproducibility and initialize the device:

```python
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
init(cuda_index=0, reproducibility=False)
```

### Main Experiment Script for Graph Coloring
This section contains the main loop for the graph coloring task, loading graph data, running the run_qubo optimization, and recording results across multiple runs.

```python

results = []  
for run_idx in range(5, 11):
    set_seed(run_idx * 10)
    graph = from_file_to_graph(Datasets.Graph_eat.path, True, True).to(get_device())
    net = DirectProbModel(graph.num_v, 100).to(get_device())
    
    gini_cons_lambda = lambda e, n: (-200 + 1 * e) / 50
    obj_cof_lambda = lambda e, n: e / 110
    obj_cons_cof_lambda = lambda e, n: e / 1000
    cons_cof_lambda = lambda e, n: e / 10
    start_time = time.time()
    loss, outs, res = run_qubo(
        type="coloring",
        net=net,
        X=torch.randn(1, 1).to(get_device()),
        graph=graph,
        num_epochs=600,
        loss_func=loss_coloring_onehot_qubo,
        lr=0.3,
        opt='AdamW',
        clip_grad=True,
        gini_cons_cof_lambda=gini_cons_lambda,
        obj_cof_lambda=obj_cof_lambda,
        cons_cof_lambda=cons_cof_lambda,
        obj_cons_cof_lambda=obj_cons_cof_lambda,
        evaluate=True
    )
    elapsed_time = round(time.time() - start_time, 2)
    print(f"Time={elapsed_time}s")
```
